```python
import pandas as pd
```

```python
df = pd.read_csv('dataNom_1Q022025.csv').query('concepto in [1, 7]')[['codigo', 'percosto', 'periodopagado', 'del', 'al', 'dias', 'cedula', 'concepto', 'vrconcepto']];
```

```python
df
```

```python
df_gtouped = df.groupby('cedula').agg({
  'codigo': 'first',
  'percosto': 'first',  # Tomar el primer valor de 'percosto'
  'periodopagado': 'first',  # Tomar el primer valor de 'periodopagado'
  'del': 'first',  # Tomar el primer valor de 'del'
  'al': 'first',  # Tomar el primer valor de 'al'
  'dias': 'sum',  # Sumar los valores de 'dias'
  'concepto': 'first',  # Tomar el primer valor de 'concepto'
  'vrconcepto': 'sum'  # Sumar los valores de 'vrconcepto'
}).reset_index();
```

```python
df_gtouped
```

```python
df_filtrado = df_gtouped[~df_gtouped['codigo'].str.startswith('ASI')]
```
